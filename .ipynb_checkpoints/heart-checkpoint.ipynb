```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
%matplotlib inline
```

Now to read and split the data into train and test sets!

```python
x = pd.read_csv("heart.csv")
y = x.iloc[:, -1:]
x = x.iloc[:, :-1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
Time to train the first SVM to classify the data!
```

```python
clf = svm.SVC()
clf.fit(x_train, y_train.values.ravel())
```

```python
count = 0.0

for i in range(270):
    try:
        if (int(clf.predict([x_test.loc[i, :]])) == int(y_test.loc[i, :])):
            print(count)
            count += 1
    except KeyError:
        print("")

print(count/54)
```

```python
lin_clf = svm.LinearSVC()
lin_clf.fit(x_train, y_train.values.ravel())
```

```python
count = 0.0

for i in range(270):
    try:
        if (int(lin_clf.predict([x_test.loc[i, :]])) == int(y_test.loc[i, :])):
            print(count)
            count += 1
    except KeyError:
        print("")

print(count/54)
```
